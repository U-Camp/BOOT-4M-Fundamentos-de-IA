```python
# Importar librerias necesarias
import matplotlib.pyplot as plt
import pandas as pd
```

Cargamos un dataset

```python
data = pd.read_csv("resources/distances_set_filtered.csv")
```

Una vez cargado para mostrar el histograma de una variable se puede hacer como

```python
data['servicios'].hist()
plt.show()
```

Para ver la correlacion entre variables se puede hacer de la siguiente manera

```python
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')

used_features = data

colormap = plt.cm.viridis
plt.figure(figsize=(12,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sb.heatmap(data.astype(float).corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```

El metodo feature selection se puede realizar utilizando SkLearn

```python
from sklearn.feature_selection import SelectKBest
```

```python
X = data.drop(['verdadero'], axis=1)
Y = data['verdadero']

best = SelectKBest(k=5)
X_new = best.fit_transform(X, Y)
X_new.shape
selected = best.get_support(indices=True)
print(X.columns[selected])
```

Finalmente, para entrenar un clasificador de NaiveBayes se puede realizar lo siguiente

```python
from sklearn.naive_bayes import GaussianNB
```

Lo ponemos a aprender con *fit()* y obtenemos predicciones para comparar

```python
# Instantiate the classifier
gnb = GaussianNB()
# Train classifier
gnb.fit(X,Y)
y_pred = gnb.predict(X)
```

Mostramos el accuracy

```python
gnb.score(X, Y)
```
